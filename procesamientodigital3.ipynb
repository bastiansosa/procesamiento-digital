```python
fs=1000
tt = np.arange(0,2,1/fs)
ss = np.cos(2*np.pi*1*tt)+np.cos(2*np.pi*12*tt)+np.cos(2*np.pi*23*tt)
plt.figure()
plt.plot(tt,ss)
SS = scipy.fft.rfft(ss)
plt.plot(ss)
ff=scipy.fft.rfftfreq(n=len(tt),d=1/fs)
```
