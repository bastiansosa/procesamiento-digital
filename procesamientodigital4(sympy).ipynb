```python
import numpy as np
import sympy as sym
import matplotlib.pylab as plt
```

GRAFICAR LA FUNCION y = COS(T)*SEN(T)+0.1 * T

```python
x = np.linspace(-10,10,1000)
t = sym.Symbol('t')
fnc = sym.cos(t)*sym.sin(2*t)+0.1*t
y = sym.lambdify(t,fnc,'numpy')(x) #con lambdify evaluo la funcion
```

```python
plt.figure()
plt.plot(x,y)
```

GRAFICAR FUNCION CON TRAMOS

s(t) = ( 0 , t<0 )

       (2(e^-t/2) , t>=0)

```python
t = sym.Symbol('t') #creo una "variable"
ft = sym.Piecewise((0,t<0) , (2*sym.exp(-t/2) , t>=0))  #la funcion piecewise separa la funcion por tramos 
```

```python
x = np.linspace(-5,5,1001) #creo lista
y = sym.lambdify(t,ft,'numpy')(x) #lambdify evalua la funcion 
dy = sym.lambdify(t,ft.diff(t,1),'numpy')(x) # (t,1) -> t = variable a derivar ; 1 -> primera derivada. .diff-> para calcular derivadas
```

```python
plt.figure()
plt.plot(x,y,label='y')
plt.plot(x,dy,label='dy')
plt.legend()
```

AQUI TRABAJE CON OTRA FUNCION POR TRAMOS A GRAFICAR

```python
T = 2*np.pi #periodo
t = sym.Symbol('t') #creacoin de variables
ft = sym.Piecewise( (-1,t < -T/2) , (-1,t<-T/4) , (1,t<T/4) , (-1,t<T/2) , (-1,True) ) #declaramos la funcion por tramos con PIecewise
```

```python
x = np.linspace(-T/2,T/2,1001) #creo la lista de numeros
y = sym.lambdify(t,ft,'numpy')(x) #evalua la funcion 
```

```python
plt.figure()
plt.plot(x,y)
```

AQUI OTRA FUNCION

cm = 1/p  -T/2 , T/2 ( s(t)*cos (2 * pi * m * t / P )dt)

```python
m = sym.Symbol('m')
fcos = sym.cos( (T*m*t) / T )
f_in = ft * fcos
cm = sym.integrate(f_in, (t,-T/2,T/2)) / T
```

```python
c0 = cm.subs(m,5)
c0
```

```python
aux=0
for i in range(7):
  c_aux = cm.subs(m,i)
  aux = c_aux * sym.cos(2*np.pi*i*t)+aux

x = np.linspace(-5,5,100)
y = sym.lambdify(t,aux,'numpy')(x)
yaux = sym.lambdify(t,ft,'numpy')(x)
plt.figure()
plt.plot(x,y,label='aprox')
plt.plot(x,yaux,label='origen')
plt.legend()
```
