```python
import numpy as np
import matplotlib.pylab as plt
import skimage.io as io
```

```python
m = np.zeros((20,20))
m[10:20,:20]=1
m
```

```python
plt.figure(figsize=(12,12))
plt.imshow(m, cmap="gray")
```

```python
plt.figure(figsize=(12,12))
plt.hist(m.ravel())
```

```python
m2 = np.zeros((20,20))
for i in range(20):
  m2[i][i]=1
plt.figure(figsize=(12,12))
plt.hist(m2.ravel())
```
